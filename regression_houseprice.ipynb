```python
# import pckages
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_squared_error
import pandas as pd
```

```python
# load data
boston = load_boston()

# randomly divide the data into training data(90%) and testing data(10%)
x_train, x_test, y_train, y_test = train_test_split(boston["data"], boston["target"], test_size = 0.1)
```

```python
# Build Decision Tree Model
reg = DecisionTreeRegressor(max_depth=4) # The teacher set the parameter by seeing the graphics
reg.fit(x_train, y_train)

# Test
pre = reg.predict(x_test)
y_pre = reg.predict(x_test)
(1-r2_score(y_test, y_pre)), mean_squared_error(y_test, y_pre), (mean_squared_error(y_test, y_pre)/(1-r2_score(y_test, y_pre)))
```
