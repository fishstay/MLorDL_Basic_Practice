```python
from keras.datasets.mnist import load_data
# ((x_train, y_train), (x_test, y_test))
(x_train, y_train), (x_test, y_test) = load_data()
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import random
i = random.randint(0, 59999)
plt.imshow(x_train[i], cmap = "gray")
plt.title("This is {}".format(y_train[i]))
```

```python
import pandas as pd
pd.DataFrame(x_train[i])
```

```python
x_train_reshape = x_train.reshape(x_train.shape[0], -1)
x_test_reshape = x_test.reshape(x_test.shape[0], -1)
a = pd.DataFrame(x_train_reshape)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 30)
pca.fit(x_train_reshape)
x_train_pca = pca.transform(x_train_reshape)
x_test_pca = pca.transform(x_test_reshape)
```

```python
pd.DataFrame(x_train_pca)
```

```python
axis = pca.components_
width = 5
height = len(axis) // width + 1
plt.figure(figsize = (14, 14))
for i, a in enumerate(axis):
    plt.subplot(height, width, i+1)
    a_reshape = a.reshape(28, 28)
    plt.axis("off")
    plt.title("New Axis:{}".format(i))
    plt.imshow(a_reshape, cmap = "gray")
```

```python
print("每一個軸解釋了多少cov", pca.explained_variance_ratio_)
print("SUM", sum(pca.explained_variance_ratio_))
```

```python
print("留下來的影響力：\n", pca.singular_values_)
```

```python
# random forest
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.model_selection import cross_val_score
import numpy as np

clf = RandomForestClassifier(n_estimators = 100, max_depth = 20, n_jobs = -1)
scores = cross_val_score(clf, x_train_pca, y_train, cv= 10, n_jobs = -1)
print(np.average(scores))

clf.fit(x_train_pca, y_train)
pre = clf.predict(x_test_pca)

accuracy = metrics.accuracy_score(y_test, pre)
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix
i = ["{}(真實)".format(n) for n in list(range(10))]
c = ["{}(預測)".format(n) for n in list(range(10))]
pd.DataFrame(confusion_matrix(y_test,pre), index = i, columns = c)
```

```python
percents = []
for r in confusion_matrix(y_test, pre):
    percents.append(["{}%".format(round(n/sum(r)*100 ,2)) for n in r])

pd.DataFrame(percents, index = i, columns = c)
```
